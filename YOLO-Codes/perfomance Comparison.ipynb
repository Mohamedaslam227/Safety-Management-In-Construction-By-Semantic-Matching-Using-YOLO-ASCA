```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
asca_file = r"E:\My Research Project\CODE\YOLO-Codes\yolo_esca_results\yolo_esca_run\results.csv"  # Replace with your file path
v11_file = r"E:\My Research Project\CODE\YOLO-Codes\yolo_esca_results\yolo_esca_run3\results.csv"    # Replace with your file path

asca_df = pd.read_csv(asca_file)
v11_df = pd.read_csv(v11_file)

# Metrics to compare
metrics_of_interest = [
    "metrics/precision(B)", "metrics/recall(B)", "metrics/mAP50(B)", "metrics/mAP50-95(B)",
    "val/box_loss", "val/cls_loss", "val/dfl_loss"
]

# Plotting function
def plot_metrics(asca_df, v11_df, metrics):
    for metric in metrics:
        plt.figure(figsize=(8, 5))
        
        # Plot YOLO ASCA
        plt.plot(asca_df['epoch'], asca_df[metric], label="YOLO ASCA", marker='o')
        
        # Plot YOLO V11
        plt.plot(v11_df['epoch'], v11_df[metric], label="YOLO V11", marker='s')
        
        # Configure plot
        plt.title(f"Comparison of {metric}")
        plt.xlabel("Epoch")
        plt.ylabel(metric)
        plt.legend()
        plt.grid(True)
        plt.show()

# Call the plotting function
plot_metrics(asca_df, v11_df, metrics_of_interest)
```
